```python
import json

def loadData(fileName):
    try:
        with open(fileName, "r") as f:
            return json.load(f)
    except FileNotFoundError:
        print("File not found:", fileName)
    except json.JSONDecodeError as e:
        print("Invalid JSON:", e)
```

```python
data=loadData("store_data.json")
```

```python
def cleanData(data):
    textToNum={"one":1,"two":2,"three":3,"four":4,"five":5}
    for user in data:
        rawData=user["rating"].strip().lower()
        if (rawData in textToNum):
            rawData=textToNum[rawData]
        user["rating"]=rawData
        print(user["rating"])
   
```

```python
cleanData(data)
```

```python
def handleAge(data):
    for user in data:
        rawage=user.get("age")
        if(rawage==None):
            user["age"]=None
        print(user["age"])
        
```

```python
handleAge(data)
```

```python
def duplicateData(data):
    unique=set()
    cleanData=[]
    for user in data:
        if(user["name"].strip() in unique):
            continue
        unique.add(user["name"])
        cleanData.append(user)
    return cleanData
    
```

```python
print(duplicateData(data))
```

```python

def insights(data):
    poorRating=0
    total=0
    for user in data:
        total+=float(user["rating"])
        if(float(user["rating"])<3.0):
            poorRating+=1
    print(f"Avg rating: {total/len(data)}")
    print(f"poor rating:{(poorRating/len(data))*100}")
```

```python
insights(data)
```
