```python
import matplotlib.pyplot as plt

age = [10,16,85,96,45,65,85,25,35,64,55]
bp = [74,79,150,155,110,120,135,95,100,105,102]

plt.scatter(age, bp, marker="<", c=bp, cmap="OrRd", label="Patients")

plt.xlabel("Age")
plt.ylabel("Blood Pressure")

plt.colorbar(label="Blood Pressure")
plt.legend()
plt.ylim(min(bp),max(bp)+10)

for i in range(len(bp)):
    plt.annotate(age[i],xy=(age[i],bp[i]),xytext=(age[i],bp[i]+2))
plt.show()
```
