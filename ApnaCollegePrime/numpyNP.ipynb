```python
import numpy as np
import time
```

```python
a=1000000
py_list=list(range(a))
start=time.time()
ls=[n**2 for n in py_list]
end=time.time()
print(f"time taken: {end-start} second")

np_array=np.array(py_list)
start=time.time()
sq_array=np_array**2
end=time.time()
print(f"time taken: {end-start} second")
```

```python
import sys
```

```python
print(f"Size of python list object: {sys.getsizeof(ls)} bytes")
print(f"Size of numpy array: {sq_array.nbytes} bytes")
```
